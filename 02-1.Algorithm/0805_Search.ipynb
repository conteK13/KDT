```python
#Binary Search를 함수로 구현
import random

#랜덤한 검색범위 data를 만들고 출력하는 함수.
def initdata():
    data = random.sample(range(1, 101), 50)       #중복없음 = 비복원 추출.
    print("[원본데이터]")
    print(data)
    
    return data

#입력받은 리스트를 정렬하고 출력하는 함수.
def sortData(data):
    Sdata = sorted(data)
    print("[정렬데이터]")
    print(Sdata)
    
    return Sdata


#정렬되지 않은 상태 또는 정렬된 상태에서 값 찾기.
def SequentialSerch(data, key):
    i = 0
    while (i< len(data) and data[i] != key):
        print("%2d번째 검색입니다." %(i+1))
        i+=1 
        
    if i<len(data):
        print("%2d번째 검색입니다." %(i+1))  
        print("원하는 값 %d는 인덱스 %d에 위치하고 있습니다." %(key, i))
        return
    else:
        print("원하는 값 %d는 원본데이터에 존재하지 않습니다." %(key))


#2진 검색에 사용하는 함수 
def BinarySerch(data, key): #a는 list형태의 검색 범위, key는 검색하고 싶은 값.
    start = 0
    end = len(data)-1       #index 설정을 위해 -1 해준다.

    count = 1
    while start <= end:
        middle = int((start+end) * 0.5)
        print(f"{count:2}번째 검색 : start={start:2}, end={end:2}, middle={middle:2}, data[{middle:2}]={data[middle]:2}")
        
        if data[middle] == key:
            print("원하는 값 %d는 인덱스 %d에 위치하고 있습니다." %(key, middle))
            return
        elif data[middle] < key:
            start = middle+1
        else:
            end = middle -1
        count += 1
    print("원하는 값 %d는 원본데이터에 존재하지 않습니다." %(key))
    return False

def binary_rec(a, start, end, key):
    global count
    count+=1
    if end < start:
        print("원하는 값 %d는 원본데이터에 존재하지 않습니다." %(key))
        return
    middle = int((start+end)*0.5)
    print(f"{count:2}번째 검색 : start={start:2}, end={end:2}, middle={middle:2}, a[{middle:2}]={a[middle]:2}")
    if a[middle] == key:
        print("원하는 값 %d는 인덱스 %d에 위치하고 있습니다." %(key, middle))
        return
    elif a[middle] <key:
        binary_rec(a, middle+1, end, key)
    else:
        binary_rec(a, start, middle-1, key)

print("serch algorithm")
print("-"*50)

data = initdata()
Sdata = sortData(data)

key = int(input("찾고자 하는 값을 입력하세요(1~100) : "))

print("-"*50)
print("SequentialSerch_unsorted")
SequentialSerch(data, key)   #정렬되지 않은 Data에 SequentialSerch

print("-"*50)
print("SequentialSerch_sorted")
SequentialSerch(Sdata, key)  #정렬된 Data에 SequentialSerch

print("-"*50)
print("BinarySerch")
BinarySerch(Sdata, key)      #정렬된 Data에 BinarySerch

print("-"*50)
print("BinarySerch : 재귀함수")
count=0
binary_rec(Sdata, 0, len(Sdata)-1, key)
```
