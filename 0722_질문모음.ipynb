```python
# 두 개의 숫자를 입력받습니다. 띄어쓰기로 구분합니다.
numbers_str = input('두 숫자 a, b를 띄어쓰기로 구분하여 입력하세요 (예: 10 20): ')

# 입력받은 문자열을 띄어쓰기를 기준으로 나눠 리스트로 만듭니다.
numbers_list = numbers_str.split()

# map 함수를 사용하여 각 숫자 문자열을 정수로 변환합니다.
numbers_int = map(int, numbers_list)

# 결과를 리스트로 변환합니다.
numbers = list(numbers_int)

# 입력받은 두 숫자를 변수 a, b에 할당합니다.
a = numbers[0]
b = numbers[1]

# 두 숫자를 더합니다.
sum = a + b

# 결과를 출력합니다.
print(f"a = {a}, b = {b}, a + b = {sum}")
```
