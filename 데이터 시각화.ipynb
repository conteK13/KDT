```python
plt.title("여러 개의 그래프 그리기")

# 0.2 간격으로 균일하게 샘플된 시간
# Matplotlib에서는 일반적으로 NumPy 어레이를 이용
t = np.arange(0., 5., 0.2)
# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
plt.title("딕셔너리 데이터 사용하기")

# 딕셔너리와 같이 레이블이 있는 데이터를 그래프로 나타낼 수 있음
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}
plt.plot( 'data_x', 'data_y', 'r--', data=data_dict) #순서 중요
#data 파라미터에 딕셔너리 지정
plt.show()
```

```python
plt.title("축 레이블 설정")

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
# 기본 사용 - fill_between() / 그래프 영역 채우기

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=1)
#  (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워짐

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)
#  (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역이 채워짐

plt.show()
```

```python
# 두 그래프 사이 영역 채우기

x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[0:3], y1[0:3], y2[0:3], color='green', alpha=0.5)
# (x[0], y1[0]), (x[2], y1[2]), (x[2], y2[2]), (x[2], y2[0]) 사이 영역을 채움
plt.show()
```

```python
# 수평선/수직선 표시

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=4)
# axhline(y값, xmin, xmax) /xmin, xmax는 0에서 1사이 값. 
# 0은 왼쪽 끝, 1은 오른쪽 끝
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
# hlines(y, xmin, xmax) / 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

plt.show()
```

```python
# 수직선 그리기

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
# axvline(x 값, ymin, ymax) 
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)
# vlines(x, ymin, ymax) / 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시
plt.show()
```

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2 #사이즈, size**2 의 형태로 지정
colors = np.random.rand(n) #색상설정

plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
# 수평 박스 플롯

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

# vert 파라미터를 False로 지정하면 수평 방향의 박스 플롯이 나타남
ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5, vert=False)
ax.set_xlim(-10.0, 10.0)
ax.set_xlabel('Value')
ax.set_ylabel('Data Type')

plt.show()
```

```python
#ax.violinplot() 옵션 사용 바이올린 형태의 플롯 시각화

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], positions=[2, 3, 4])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# showmeans 파라미터는 데이터 분포에서 평균값의 위치에 직선을 표시(Default: False)
# showextrema는 최대/최소값에 직선을 표시(Default: True)
# showmedians는 중간값에 직선을 표시(Default: False)

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

# 색상 지정
violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

plt.show()
```
