### 7) 여러개의 그래프

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

#plt.subplot(3, 1, 2)
# 3행 1열의 2번째 위치를 내가 쓰겟다
# subplot의 행과 열의 값은 동일해야한다.

#subplot 위치들만 확인
plt.subplot(3, 1, 2)                # nrows=3, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(3, 1, 1)                # nrows=3, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.subplot(3, 1, 3)                # nrows=3, ncols=1, index=3
plt.plot(x2, y2, '.-')
plt.title('3nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(1, 3, 1)                # nrows=1, ncols=2, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time (s)')
plt.ylabel('Damped oscillation')

plt.subplot(1, 3, 2)                # nrows=1, ncols=2, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.subplot(1, 3, 3)                # nrows=1, ncols=2, index=3
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

### 11) 두 종류의 그래프 그리기

```python
# 1. 기본 스타일 및 설정
plt.style.use('default')  # 기본 스타일 사용
plt.rcParams['figure.figsize'] = (4, 3)  # 그림 크기 설정
plt.rcParams['font.size'] = 12  # 폰트 크기 설정
# 그래프의 기본 스타일, 폰트, 색상, 축의 설정 등 다양한 속성을 한 곳에서 설정할 수 있음.

# 2. 데이터 생성
x = np.arange(2020, 2027)  # x축 데이터: 2020년부터 2026년까지
y1 = np.array([1, 3, 7, 5, 9, 7, 14])  # y1 데이터: 가격
y2 = np.array([1, 3, 5, 7, 9, 11, 13])  # y2 데이터: 수요

# 3. 그래프 그리기
fig, ax1 = plt.subplots()  # 새로운 서브플롯 생성

# 꺾은선 그래프를 ax1에 그림
ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0, 18)  # y축 범위 설정
ax1.set_xlabel('Year')  # x축 레이블 설정
ax1.set_ylabel('Price ($)')  # y축 레이블 설정

# 공유 y축을 가지는 두 번째 y축 생성
ax2 = ax1.twinx()  # ax1의 두 번째 y축 생성

# 막대 그래프를 ax2에 그림
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0, 18)  # y축 범위 설정
ax2.set_ylabel(r'Demand ($\times10^6$)')  # 두 번째 y축 레이블 설정

# 그래프 표시
plt.show()
```

fig, ax1 = plt.subplots()

1. fig
타입: matplotlib.figure.Figure
설명: fig는 matplotlib의 Figure 객체를 참조합니다. Figure 객체는 전체 그림을 캔버스처럼 다루는 역할을 합니다. 여기에는 하나 이상의 서브플롯(Axes), 제목, 레이블, 전체적인 스타일과 레이아웃 등을 포함합니다.

주요 역할:
그림 전체의 크기와 모양을 설정합니다.
하나 이상의 서브플롯을 포함합니다.
그림에 대한 다양한 설정을 조정할 수 있습니다 (예: 전체 제목, 저장, 크기 조정 등).


2. ax1
타입: matplotlib.axes.Axes
설명: ax1은 Figure 객체에 포함된 Axes 객체를 참조합니다. Axes 객체는 실제 그래프나 플롯이 그려지는 영역을 의미합니다. 이 객체는 데이터의 시각화와 관련된 여러 메서드를 제공하며, x축, y축, 축 레이블, 제목 등을 설정합니다.

주요 역할:
데이터를 실제로 플로팅합니다 (예: 꺾은선 그래프, 막대 그래프 등).
x축과 y축의 레이블 및 한계를 설정합니다.
그래프의 제목, 축의 눈금, 스타일 등을 설정합니다.
