### 5) 파이 차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
# startangle 시작각도 설정(디폴트 0도는 3시 방향임, 반시계뱡향으로 degree를 조절 )
# counterclock=False로 설정하면 시계 방향 순서로 부채꼴 영역이 표시됨

plt.show()
```

```python
# 시작 각도와 방향 설정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0, 0, 0.20]
# explode
# 0 에 가깝다면 중심에서 떨어져 나가지 x, 값이 클수록 분리된 정도가 커짐. 
# 1이 되면 반지름 만큼 떨어져나감, 최대값에 제한이 없음.

# autopct 옵션 : 조각의 비율을 숫자와 %의 조합으로 표시해줌
plt.pie(ratio, labels=labels, autopct='%.2f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

```python
# 중심에서 벗어나는 정도 설정 / 색상 지정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]
# explode는 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정
# Banana’와 ‘Grapes’ 영역에 대해서 반지름의 10% 만큼 벗어나도록 설정
colors = ['silver', 'gold', 'lightblue', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
#shadow를 True로 설정하면, 파이 차트에 그림자가 표시
plt.show()
```

```python
# 부채꼴 스타일 지정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.5, 'edgecolor': 'w', 'linewidth': 2}
# ‘wedgeprops' 딕셔너리를 이용하여 옵션 설정
# 지정할 수 있는 옵션 :
# edgecolor: 웨지의 테두리 색상을 설정합니다.
# linewidth: 웨지의 테두리 두께를 설정합니다.
# linestyle: 웨지 테두리의 선 스타일을 설정합니다.
# antialiased: 안티앨리어싱을 활성화할지 여부를 설정합니다.

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)

plt.show()
```
