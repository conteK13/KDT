```python
# date_added: 콘텐츠가 데이터베이스에 추가된 날짜입니다.
net['date_added'] = net['date_added'].str.strip() #앞뒤 공백 제거  중간에 공백이 있는 부분이 있어서

# 'date_added'에서 달만 추출
net['month_added'] = pd.to_datetime(net['date_added']).dt.strftime('%B')#날짜 데이터들을 문자열에서 날짜 형식으로 변환 후 dt.strftime을 이용해 월만 추출

# 달별 작품 수 집계
monthly_counts = net['month_added'].value_counts().reindex([ #value_counts()는 기본적으로 값이 많은 순서대로 정렬됩니다. 
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# 그래프 시각화
monthly_counts.plot(kind='bar', color='yellow', figsize=(8, 5)) 
plt.title('month contents')
plt.xlabel('Month')
plt.ylabel('Number of content')
plt.xticks(rotation=45) #레이블을 기울여 글자가 겹치는 현상을 막아줌
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 'date_added'에서 월만 추출
net['month_added'] = pd.to_datetime(net['date_added']).dt.strftime('%B')

# 월별 및 type별 콘텐츠 수 집계
monthly_type_counts = net.groupby(['month_added', 'type']).size().unstack().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# 그래프 시각화
monthly_type_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Monthly Content Count by Type')
plt.xlabel('Month')
plt.ylabel('Number of Content')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```
