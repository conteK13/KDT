### 2) 막대 그래프

```python
x = np.arange(3)
# range로 인덱스를 잡아준다(0부터 2020까지 그리면 데이터 시각화 할때 문제가 생길 수 있으므로)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
#x축의 눈금 레이블에 '2018', '2019', '2020'이 순서대로 표시
# 실제 값 변경 x. 레이블만 변경해줌

plt.show()
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y','dodgerblue', 'C2']

# plt.bar(x, values, width = 0.4)

# plt.bar(x, values, width = 0.6, color=colors)
plt.bar(x, values, width = 0.8, color=colors)
# plt.bar(x, values, width = 1.0, color=colors)
plt.xticks(x, years)

plt.show()
```

```python
x = np.arange(4)
years = ['2018', '2019', '2020', '2021']
values = [100, 400, 900,500]
colors = ['y','dodgerblue']

# plt.bar(x, values, width = 0.4, color=colors)
# plt.bar(x, values, width = 0.6, color=colors)
plt.bar(x, values, width = 0.8 , color=colors)
# plt.bar(x, values, width = 1 , color=colors)
plt.xticks(x, years)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(10)
years = [f'Year {i+1}' for i in x]  # 예시로 x축에 표시할 연도
values = np.random.randint(100, 1000, size=10)  # 임의의 값

# 기본 색상 리스트 생성
base_colors = ['yellow', 'dodgerblue', 'purple']

# 색상 리스트 생성: 색상이 반복되도록 함
colors = [base_colors[i % len(base_colors)] for i in range(10)]

# 5번째 막대(인덱스 4)의 색상만 빨간색으로 변경
colors[4] = 'red'

# 막대그래프 그리기
plt.bar(x, values, color=colors)
plt.xticks(x, years)

plt.xlabel('Years')
plt.ylabel('Values')
plt.title('색상 옵션')

plt.show()
```

```python
# 스타일 꾸미기

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, align='edge', edgecolor='lightgray',
        linewidth=5, tick_label=years, hatch='\\')
# {'/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*'} 채우기 옵션

plt.show()
```

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values)
plt.yticks(y, years)
#bar+h 함으로서 y축을 기준으로 그래프를 그린다.
# Y축의 눈금 레이블에 ‘2018’, ‘2019’, ‘2020’이 순서대로 표시

plt.show()
```

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

# plt.barh(y, values, height= 0.4)
plt.barh(y, values, height = 0.6)
# plt.barh(y, values, height = 0.8)
# plt.barh(y, values, height = 1.0)
plt.yticks(y, years)
#bar+h 함으로서 y축을 기준으로 그래프를 그린다.
# Y축의 눈금 레이블에 ‘2018’, ‘2019’, ‘2020’이 순서대로 표시

plt.show()
```
