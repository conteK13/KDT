```python
# 선 종류 지정
#‘ - ‘ (Solid), ‘ - - ‘ (Dashed), ‘ : ‘ (Dotted), ‘ -. ‘ (Dash-dot)

plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4, fontsize=8)
plt.show()

#선 모양을 범례에 넣는 방법은 plt.plot()에서 label 매개변수를 설정하고, plt.legend()를 호출하여 그 내용을 범례로 표시하는 것입니다.
#범례에 표시되는 내용은 각 선의 label 속성에 따라 자동으로 결정됩니다.

"""
C0, C1, C2, C3 등은 Matplotlib에서 자동으로 할당되는 기본 색상 팔레트의 색상을 나타내는 값입니다. 이 색상들은 Matplotlib이 그래프에 색을 지정할 때 사용하는 순차적인 색상 배열에서 가져오며, 기본 설정으로 제공되는 색상입니다.

Matplotlib에서 기본 색상 팔레트는 10가지 색상으로 이루어져 있으며, 그 값들은 고정되어 있습니다. 이 값들은 다음과 같이 순서대로 할당됩니다:

C0: 파란색 (blue)
C1: 주황색 (orange)
C2: 녹색 (green)
C3: 빨간색 (red)
C4: 보라색 (purple)
C5: 갈색 (brown)
C6: 분홍색 (pink)
C7: 회색 (gray)
C8: 올리브색 (olive)
C9: 하늘색 (cyan)
기본적으로 10개의 색상이 제공되며, 더 많은 데이터 시리즈가 있을 경우 이 색상들이 다시 반복됩니다.
"""
```

```python
#‘ - ‘ (Solid), ‘ - - ‘ (Dashed), ‘ : ‘ (Dotted), ‘ -. ‘ (Dash-dot)

plt.figure(figsize=(10,5 ))
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

```python
# 선 끝 모양 지정

plt.figure(figsize=(10, 6))
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=5,
      solid_capstyle='butt', color='r', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')
# dash_capstyle를 사용해서 선의 끝 모양 지정
# {'butt', 'projecting', 'round'}

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
# 기본 사용 - fill_between() / 그래프 영역 채우기

x = [1,2,3,4]
y = [2,3,5,10]
plt.plot(x,y)

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], color = "red", alpha = 1)
# x[1:3]은 x의 1번째 인덱스부터 2번째 인덱스까지
# y[1:3]은 y의 1번째 인덱스부터 2번째 인덱스까지
# alpha : 투명도(0이면 완전 투명, 1이면 완전 불투명. alpha는 반드시 0과 1사이의 값을 갖는다.)

plt.show()
```

```python
# 수평선/수직선 표시

x = np.arange(0, 4, 0.5)

# plt.plot(x, x**1, 'bo')
# plt.plot(x, x**2, 'g--')
# plt.plot(x, x**3, 'r:')

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

#v와 h는 각각 "vertical"과 "horizontal"의 약자
plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=4)
# axhline(y값, xmin, xmax) /xmin, xmax는 0에서 1사이 값. 0은 왼쪽 끝, 1은 오른쪽 끝
# x값이 절대값이 아님

plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
# hlines(y, xmin, xmax) / 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

plt.show()
```

```python
# 수직선 그리기

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

#v와 h는 각각 "vertical"과 "horizontal"의 약자
plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
# axvline(x 값, ymin, ymax) 
# y값이 절대값이 아니라 비율임

plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)
# vlines(x, ymin, ymax) / 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시
plt.show()
```
