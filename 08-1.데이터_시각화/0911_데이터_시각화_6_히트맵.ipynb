### 6) 히트맵

```python
arr = np.random.standard_normal((30, 40))
# 표준정규분포를 갖는 (30, 40) 형태의 2차원 어레이 
# 30행, 40열

plt.matshow(arr)
plt.colorbar(shrink=0.8, aspect=10) # 컬러바 추가
# shrink: 컬러바의 크기를 조정합니다. 0.8은 기본 크기의 80%로 축소합니다.
# aspect: 컬러바의 종횡비를 설정합니다. 기본값은 20이며, 값이 낮을수록 컬러바가 더 넓어집니다.

# plt.clim(-1.0, 1.0)
# 이 줄의 주석을 해제하면 색상 범위를 -1.0에서 1.0으로 설정합니다.

plt.show()
```

```python
# 색상 범위 지정

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink=0.8, aspect=10)
#plt.clim(-1.0, 1.0)
# 색상의 범위를 -1.0 ~ 1.0 으로 지정, arr의 값 중 -1.0 보다 작거나 1.0 보다 큰 값에 대해서는 각각 -1.0, 1.0과 같은 색으로 나타냄
#오른쪽 컬러맵의 범위가 변경되는것을 확인할 수 있다. 
plt.clim(-2.0, 2.0)

plt.show()
```

```python
#색상 변경하기
arr = np.random.standard_normal((30, 40))
#cmap = plt.get_cmap('PiYG')
#cmap = plt.get_cmap('BuGn')
cmap = plt.get_cmap('Greys')
# cmap = plt.get_cmap('bwr')

plt.matshow(arr, cmap=cmap)
#  get_cmap() 함수를 이용해서 Matplotlib 컬러맵을 가져와서 matshow()에 입력
plt.colorbar()
plt.show()
```
