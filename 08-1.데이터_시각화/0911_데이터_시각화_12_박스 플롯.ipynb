### 12) 박스 플롯

일반적으로 박스 플롯은 전체 데이터로부터 얻어진 아래의 다섯 가지 요약 수치를 사용해서 그려짐

- 최소값
- 제 1사분위 수 (Q1)
- 제 2사분위 수 또는 중위수 (Q2)
- 제 3사분위 수 (Q3)
- 최대값

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 및 설정
plt.style.use('default')  # 기본 스타일 적용
plt.rcParams['figure.figsize'] = (4, 3)  # 그림의 크기를 가로 4인치, 세로 3인치로 설정
plt.rcParams['font.size'] = 12  # 폰트 크기를 12로 설정

# 2. 데이터 생성
np.random.seed(0)  # 난수 생성 시드를 설정하여 결과의 재현성을 보장
data_a = np.random.normal(0, 2.0, 1000)  # 평균 0, 표준편차 2.0인 정규분포에서 1000개의 샘플 생성
data_b = np.random.normal(-3.0, 1.5, 500)  # 평균 -3.0, 표준편차 1.5인 정규분포에서 500개의 샘플 생성
data_c = np.random.normal(1.2, 1.5, 1500)  # 평균 1.2, 표준편차 1.5인 정규분포에서 1500개의 샘플 생성

# 3. 박스 플롯 생성
fig, ax = plt.subplots()  # 새로운 Figure와 Axes 객체 생성

ax.boxplot([data_a, data_b, data_c])  # data_a, data_b, data_c의 박스 플롯을 생성
ax.set_ylim(-10.0, 10.0)  # y축의 범위를 -10.0에서 10.0으로 설정
ax.set_xlabel('Data Type')  # x축 레이블을 'Data Type'으로 설정
ax.set_ylabel('Value')  # y축 레이블을 'Value'로 설정

plt.show()  # 생성된 플롯을 화면에 표시
```

```python
# 중앙값 표나게 노치 표시하기

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5)
# notch : 상자 위 아래에 잇는 음푹 들어간 부분을 추가하여 중앙값의 신뢰구간을 시각적으료 표현하는 옵션
# whis : 수염(whiskers)은 데이터의 분포를 시각적으로 표현하는 요소
# 길이를 조절함으로서 데이터의 범위를 표시한다.

ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# 수평 박스 플롯

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

# vert 파라미터를 False로 지정하면 수평 방향의 박스 플롯이 나타남
ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5, vert=False)
ax.set_xlim(-10.0, 10.0)
ax.set_xlabel('Value')
ax.set_ylabel('Data Type')

plt.show()
```

```python
#ax.violinplot() 옵션 사용 바이올린 형태의 플롯 시각화

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], positions=[2, 3, 4])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# showmeans 파라미터는 데이터 분포에서 평균값의 위치에 직선을 표시(Default: False)
# showextrema는 최대/최소값에 직선을 표시(Default: True)
# showmedians는 중간값에 직선을 표시(Default: False)

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

# 색상 지정
violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

plt.show()
```
