### 0) 기본적인 기능들 - 1. 텍스트 표시

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(1.0, 0.35, '2.0*np.random.randn(10000)+1.0')
# text(x, y, '문구')
# x, y의 위치에 '문구'를 출력한다
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.20, 'np.random.standard_normal(10000)')

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(5.0, 0.08, 'np.random.rand(5000)-10.0')
plt.show()
```

```python
#fontdict 키워드 이용

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family': 'serif',
      'color':  'darkred',
      'weight': 'normal',
      'size': 16}
# font의 종류, 크기, 색상, 투명도, weight 등의 텍스트 스타일을 설정 가능

font2 = {'family': 'Times New Roman',
      'color':  'blue',
      'weight': 'bold',
      'size': 12,
      'alpha': 0.7}

font3 = {'family': 'Arial',
      'color':  'forestgreen',
      'style': 'italic',
      'size': 14}

"""
fontdict 옵션
family: 폰트 패밀리 (예: 'serif', 'sans-serif', 'monospace')
weight: 폰트의 두께 (예: 'normal', 'bold', 'light')
size: 폰트 크기 (예: 12, 'large')
color: 폰트 색상 (예: 'red', 'blue', '#FFDD44')
style: 폰트 스타일 (예: 'normal', 'italic', 'oblique')
variant: 폰트 변형 (예: 'normal', 'small-caps')
"""


#plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
#plt.text(1.0, 0.35, 'np.random.randn()', fontdict=font1)

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.20, 'np.random.standard_normal()', fontdict=font3)

#plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
#plt.text(5.0, 0.08, 'np.random.rand()', fontdict=font3)

plt.show()
```

```python
font1 = {'family': 'serif',
      'color':  'darkred',
      'weight': 'normal',
      'size': 16}

# x좌표와 y좌표 0,0 은 본래 좌측 하단을 기준으로 함
# 데이터의 분포에 따라 x와 y 축의 범위가 설정되는데, 이때 데이터의 밀도 분포에 맞춰 조정
# 때문에 text의 위치 (0, 0)가 플롯의 시각적 가운데 하단에 나타날 수 있음.

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(0, 0, 'test', fontdict=font1)

plt.show()
```

```python
# 텍스트 회전 / rotation를 이용해 첫번째, 두번째 텍스트를 각각 85도, -60도만큼 회전

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(-3.0, 0.15, 'np.random.randn()', fontdict=font1, rotation=85)
#rotation은 회전각 설정

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict=font2, rotation=-60)

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict=font3, rotation=90)
#시계 반시계방향으로 회전한다.

plt.show()
```

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16}

font2 = {'family': 'Times New Roman',
        'color':  'blue',
        'weight': 'bold',
        'size': 12,
        'alpha': 0.7}

font3 = {'family': 'Arial',
        'color':  'forestgreen',
        'style': 'italic',
        'size': 14}

box1 = {'boxstyle': 'round',
        'ec': (1.0, 0.5, 0.5),
        'fc': (1.0, 0.8, 0.8)}

box2 = {'boxstyle': 'square',
        'ec': (0.5, 0.5, 1.0),
        'fc': (0.8, 0.8, 1.0),
        'linestyle': '--'}

#이런 방식이 있구나 까지만 생각할것. 자주 사용x
box3 = {'boxstyle': 'square',
        'ec': (0.3, 1.0, 0.5), #‘ec’는 ‘edgecolor’
        'fc': (0.8, 1.0, 0.5), #‘fc’는 ‘facecolor’
        'linestyle': '-.',
        'linewidth': 2}

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(-3.0, 0.15, 'np.random.randn()', fontdict=font1, rotation=85, bbox=box1)
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict=font2, rotation=-60, bbox=box2)
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict=font3, bbox=box3)
plt.show()
```

### 0) 기본적인 기능들 - 2. 이미지 저장

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
# plt.show()
plt.savefig('savefig_default.png')
#plt.savefig('resource/savefig_default.png')
#같은 폴더의 resource 폴더안에 저장
```
