### 4) 산점도

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x,y)
plt.show()
```

시드는 난수 생성기의 초기 상태를 결정하는 값입니다.
난수 생성기는 본질적으로 결정적인 알고리즘을 사용하여 숫자를 생성하므로,
동일한 시드를 사용하면 동일한 난수 시퀀스가 생성됩니다.

이를 통해 난수 생성의 결과를 재현할 수 있습니다.

np.random.seed(0)의 역할:
결정적 결과: np.random.seed(0)을 설정하면,
이후에 생성되는 난수는 동일한 시드를 사용하는 한 동일한 값들을 생성합니다.
즉, 코드가 실행될 때마다 동일한 난수 결과를 보장합니다.

재현성: 데이터 분석이나 실험에서 결과를 재현할 수 있게 해줍니다.
예를 들어, 그래프나 모델 결과가 항상 같아야 할 때 유용합니다.

코드의 디버깅이나 결과 검증을 용이하게 만듦.


```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30*np.random.rand(n))**2 #사이즈, size**2의 형태로 지정
# 점의 크기를 보여주는 변수
# 30*np.random.rand(n) : 0부터 30 사이의 값을 출력(0과 30은 출력하지 않는다)
# **2를 함으로서 면적의 값으로 설정한다.

colors = np.random.rand(n) #색상 설정
plt.scatter(x,y, s= area, c= colors)

# s의 기본값은 20
# 컬러 맵이 0.1이면 어두운 색상, 0.9이면 더 밝은 색상임

plt.show() 
```

1. area = (30 * np.random.rand(n))**2
세부 설명:
np.random.rand(n): 0과 1 사이의 난수를 n개(여기서는 50개) 생성합니다.

이때 생성된 값들은 [0, 1) 사이의 실수입니다.
30 * np.random.rand(n): 이 값은 난수에 30을 곱하여 [0, 30) 사이의 값들을 만들게 됩니다. 즉, 각 점의 크기를 30배 범위로 확장합니다.

(30 * np.random.rand(n))**2: 그 값을 제곱하여 점의 크기가 제곱의 형태로 커지게 만듭니다. 이는 작은 차이의 값들이 더 큰 차이를 가지게 만드는 역할을 하며, 점의 크기를 더 다양하게 보이게 합니다.

결과적으로 **area**는 각 점의 크기를 결정하는 값입니다. scatter() 함수에서 이 값은 s 매개변수에 전달되어 각 점의 면적을 설정하는 데 사용됩니다.

기본 값:
plt.scatter()에서 **s(크기)**의 기본값은 s=20입니다.
즉, s 값을 지정하지 않으면 모든 점의 크기가 기본적으로 동일하게 20으로 설정됩니다.
크기를 지정하는 값은 면적을 나타내며, 값이 클수록 점이 더 커집니다.
2. colors = np.random.rand(n)
세부 설명:
np.random.rand(n): 0과 1 사이의 난수를 n개(50개) 생성합니다. 이 값은 각 점의 색상을 나타냅니다.

c=colors: scatter() 함수에 전달될 때, 각 점에 할당될 색상을 지정합니다. colors는 0과 1 사이의 값을 가지므로, 이 값은 컬러맵을 기준으로 점의 색상을 결정하게 됩니다.

기본 값:
plt.scatter()에서 **c(색상)**의 기본값은 모든 점이 동일한 색상을 가지도록 설정됩니다.
기본적으로 색상을 지정하지 않으면, 모든 점이 동일한 색상(보통 파란색)으로 표시됩니다.
색상은 기본적으로 RGB 색상으로 설정되며, 특정 컬러맵을 지정하지 않으면 Matplotlib의 기본 컬러맵을 사용하여 색상이 결정됩니다.
색상 설정 방식:
c 매개변수에 전달되는 값이 0과 1 사이의 실수일 경우, 이는 Matplotlib의 컬러맵을 참조하여 각 숫자에 대응하는 색상을 할당합니다.
기본적으로 viridis 컬러맵이 사용됩니다. 이는 낮은 값(0)에 가까울수록 어두운 색상, 높은 값(1)에 가까울수록 밝은 색상을 가지는 컬러맵입니다.
예를 들어, colors = np.random.rand(n)에서 생성된 값이 0.1이면, 컬러맵에서 어두운 색상이 선택되고, 값이 0.9이면 더 밝은 색상이 선택됩니다.


```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (10 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, alpha=1, cmap='Spectral')
# cmap='Spectral' 컬러맵에 해당하는 문자열 지정
# https://matplotlib.org/stable/gallery/color/colormap_reference.html

plt.colorbar() # 오른쪽에 컬러바 생성
plt.show()
```

```python
np.random.seed(0)
arr = np.random.standard_normal((8, 100))

plt.subplot(2, 2, 1)
#plt.scatter(arr[0], arr[1], c=arr[1], cmap='spring')

plt.scatter(arr[0], arr[1], c=arr[1])
plt.cool()
plt.title('cool')

plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c=arr[3])
plt.summer()
plt.title('summer')

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c=arr[5])
plt.autumn()
plt.title('autumn')

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c=arr[7])
plt.winter()
plt.title('winter')

# 서브플롯 간의 간격 조정
plt.tight_layout()
plt.show()
```
