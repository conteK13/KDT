### 4)히스토그램

```python
# 히스토그램 : 연속형 데이터를  여러 구간(bins)으로 나누어
# 각 구간에 포함된 데이터의 빈도를 막대 그래프로 나타내어,
# 데이터의 빈도와 분포를 나타내는 그래프

# 데이터: 각 사람의 체중 (단위: kg)
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

# 히스토그램을 그립니다.
plt.hist(weight)

# 그래프를 화면에 표시합니다.
plt.show()
```

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight, bins=10, label='bins=10') #bin(구간) 기본값 10
plt.hist(weight, bins=30, label='bins=30')
plt.legend()
plt.show()

# 지금까지 잘못 알고 있었던 개념에 대하여
# bins가 각 구간의 넓이라고 생각했는데,
# bins는 구간의 넓이가 아니라 구간의 개수를 나타냅니다
```

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight, cumulative=True, label='cumulative=True') # cumulative를 True로 두면 누적
plt.hist(weight, cumulative=False, label='cumulative=False')
plt.legend(loc='upper left')
plt.show()

#같은 데이터(weight)를 사용하는데, cumulative=True는 데이터가 누적되는 형태, flase는 그 자체의 값
```

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
        80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]


# {‘bar’, ‘barstacked’, ‘step’, ‘stepfilled’}  기본 값은 bar
plt.hist((weight, weight2), histtype='bar')
plt.title('histtype - bar')
plt.figure()

plt.hist((weight, weight2), histtype='barstacked')
plt.title('histtype - barstacked')
plt.figure()

plt.hist((weight, weight2), histtype='step')
plt.title('histtype - step')
plt.figure()

plt.hist((weight, weight2), histtype='stepfilled')
plt.title('histtype - stepfilled')
plt.show()
```

```python
# Numpy 난수 분포 나타내기

# 표준편차 2.0, 평균 1.0을 갖는 정규분포를 생성
a = 2.0 * np.random.randn(10000) + 1.0
# np.random.randn(10000)은 평균이 0이고 표준편차가 1인 정규분포를 따르는 10000개의 난수를 생성합니다.
# 여기에 2.0을 곱하고 1.0을 더하여 표준편차가 2.0, 평균이 1.0인 정규분포를 생성합니다.

# 표준정규분포를 따르는 10000개의 난수 생성
b = np.random.standard_normal(10000)
# np.random.standard_normal(10000)은 평균이 0이고 표준편차가 1인 정규분포를 따르는 10000개의 난수를 생성합니다.

# 10.0에서 -10.0 사이의 균일한 분포를 갖는 5000개의 난수 생성
c = 20.0 * np.random.rand(5000) - 10.0
# np.random.rand(5000)은 0과 1 사이의 균일한 분포를 갖는 5000개의 난수를 생성합니다.
# 이를 20.0을 곱하고 -10.0을 더하여 -10.0에서 10.0 사이의 균일 분포를 가진 데이터를 생성합니다.

# 히스토그램을 그리기
plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step', label='표준편차 2.0, 평균 1.0')
# a 데이터의 히스토그램을 그립니다. 100개의 구간으로 나누고, 밀도 기반으로 설정하여 막대의 면적이 1이 되도록 합니다.
# alpha=0.7은 투명도를 설정하며, histtype='step'은 계단 형태로 히스토그램을 그립니다.

plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled', label='표준편차 1, 평균 0 -> 표준정규분포')
# b 데이터의 히스토그램을 그립니다. 50개의 구간으로 나누고, 밀도 기반으로 설정합니다.
# alpha=0.5는 더 낮은 투명도를 설정하며, histtype='stepfilled'는 계단 형태로 채운 히스토그램을 그립니다.

plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step', label='-10.0에서 10.0 사이의 균일 분포')
# c 데이터의 히스토그램을 그립니다. 100개의 구간으로 나누고, 밀도 기반으로 설정합니다.
# alpha=0.9는 높은 투명도를 설정하며, histtype='step'은 계단 형태로 히스토그램을 그립니다.

# 그래프를 화면에 표시합니다.
plt.legend()
plt.show()
```
